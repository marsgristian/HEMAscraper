# This Scraper collects the Fiore Books

- Translation 	Michael Chidester
- Getty Version 	Michael Chidester https://wiktenauer.com/wiki/Index:Fior_di_Battaglia_(MS_Ludwig_XV_13)
- Morgan Version 	Michael Chidester https://wiktenauer.com/wiki/Index:Tratt%E2%80%8Dato_della_sch%E2%80%8Derma_(MS_M.383)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
import requests
from bs4 import BeautifulSoup


txt_table = pd.read_html("https://wiktenauer.com/wiki/Fiore_de'i_Liberi#Treatise")
print(len(txt_table))
```

```python
preface = txt_table[5]
print(preface.columns)

# preface
```

```python
# joining the prefaces into one df

preface = pd.concat([txt_table[2],txt_table[5]],ignore_index=True)
preface = preface.drop(preface.columns[[0,3,6,7,8,9,10]], axis=1)

# preface
```

```python


grappling = pd.concat([txt_table[6],txt_table[7]],ignore_index=True)
grappling = grappling.drop(grappling.columns[[0,1,3,6,7]], axis=1)

# grappling
```

```python
dagger = pd.concat([txt_table[8],txt_table[9],txt_table[10],txt_table[11],txt_table[12],txt_table[13],txt_table[14],txt_table[15],txt_table[16],txt_table[17],txt_table[18]],ignore_index=True)
dagger = dagger.drop(dagger.columns[[0,1,3,6,7,8,9]], axis=1)

dagger
```

```python
sword = pd.concat([txt_table[19],txt_table[20],txt_table[21],txt_table[22],txt_table[23], txt_table[24]],ignore_index=True)
sword = sword.drop(sword.columns[[0,1,3,6,7]], axis=1)

sword
```

```python
other_weapons = pd.concat([txt_table[25],txt_table[26],txt_table[27],txt_table[28]],ignore_index=True)
other_weapons = other_weapons.drop(other_weapons.columns[[0,1,3,6,7,8]], axis=1)

other_weapons
```

```python
# Now save'em
preface.to_csv('../Data/Fiore/preface/preface.csv',index=False)
grappling.to_csv('../Data/Fiore/grappling/grappling.csv',index=False)
dagger.to_csv('../Data/Fiore/dagger/dagger.csv',index=False)
sword.to_csv('../Data/Fiore/sword/sword.csv',index=False)
other_weapons.to_csv('../Data/Fiore/otherweapons/other_weapons.csv',index=False)
```
